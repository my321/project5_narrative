```python
import requests
import os
from pprint import pprint
import csv
import json
import time
```

### Scrape by certain date range since we can only have 100 pages worth of results (10 articls per page), and sleep for 6 seconds after every page turn, save each page in different json file since not all results have same dictionary key 


```python
apikey = os.getenv('NYTIMES_APIKEY', 'ufJL0jZynKn50qXLjAtOHxARJRhSs4en')
```

```python
#"20040101" to "20051201" 
for i in range(0,91):
    query = "Clinton, Hillary Rodham"
    begin_date = "20040101"  # YYYYMMDD
    end_date='20051201'
    filter_query = "\"body:(\"Hillary Clinton\") AND type_of_material:(\"Article\")\""  
                                            # http://www.lucenetutorial.com/lucene-query-syntax.html
    page = i  # <0-100>
    sort = "relevance"  # newest, oldest
    query_url = f"https://api.nytimes.com/svc/search/v2/articlesearch.json?" \
                f"q={query}" \
                f"&api-key={apikey}" \
                f"&begin_date={begin_date}" \
                f"&end_date={end_date}" \
                f"&fq={filter_query}" \
                f"&page={page}" \
                f"&sort={sort}"
    r = requests.get(query_url)
    keys=r.json()['response']['docs'][0].keys()
    file = r.json()['response']['docs']
        
    name_file="clinton_{num}.json".format(num=i) 
    with open(name_file, 'w') as output_file:
        json.dump(file, output_file)
    time.sleep(6)
```
